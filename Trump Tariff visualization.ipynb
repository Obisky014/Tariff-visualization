```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/stanl/Downloads/trump_tariffs_dataset.csv")
df.head()
```

```python
df["Import_Export_Ratio"] = (df["Imports"] - df["Exports"]) / df["Imports"]
```

```python
plt.figure(figsize=(12, 10))
scatter = plt.scatter(
    df["Import_Export_Ratio"],
    df["Tariff Rate"],
    c=df["Tariff Rate"],
    cmap='viridis',
    s=100,
    alpha=0.8,
    edgecolors='w'
)
for i, row in df.iterrows():
    plt.text(row["Import_Export_Ratio"],
row["Tariff Rate"], row["Country"],
fontsize=9)

cbar = plt.colorbar(scatter)
cbar.set_label('Tariff Rate')

plt.title('Trump Claimed Tariffs vs Skewed Tariff Formula')
plt.xlabel('Skewed Tariff Formula')
plt.ylabel('Claimed Tariff Rate')
plt.grid(True)
plt.tight_layout()
plt.show()
          
```
